# Analyse des Joueurs et Jeux Vidéo

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings


%load_ext autotime
warnings.filterwarnings('ignore', category=UserWarning)
```

```python
avis = pd.read_csv('../Data/avis.csv')
avis.fillna('', inplace=True)
avis.head()
```

```python
#Tout les auteurs
auteurs_uniques = avis["author"].unique()
#resultat = avis[avis["author"].isin(auteurs_uniques)]
print(f"nombre d'autheur/joueurs: {len(auteurs_uniques)}")
```

```python
# Convertir la liste des auteurs uniques en DataFrame pandas
auteurs = pd.DataFrame(auteurs_uniques, columns=["author"])

# Ajouter la colonne les colonnes avec des valeurs initiales à zéro
auteurs["nb_avis"] = 0
auteurs["notes_moyennes"] = np.nan
auteurs["ecart-type"] = np.nan
auteurs["nb_commentaires"] = 0
# Afficher les premières lignes du DataFrame avec la nouvelle colonne ajoutée
print(auteurs.head(5))
```

```python
auteurs = pd.read_csv('../Data/auteurs.csv')
auteurs.head()
```

```python
# Attempt a direct and efficient plotting method for Mean Rating vs. Standard Deviation

plt.figure(figsize=(10, 6))

# Direct scatter plot using matplotlib for more control and efficiency
plt.scatter(auteurs['ecart-type'], auteurs['notes_moyennes'], alpha=0.7, color='blue', edgecolor='black')

# Enhancing the plot
plt.title('Mean Rating vs. Standard Deviation for Each Author')
plt.xlabel('Standard Deviation of Ratings')
plt.ylabel('Mean Rating')
plt.grid(True)  # Adding grid for better readability

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Density Heatmap for Mean Rating vs. Standard Deviation
sns.kdeplot(data=auteurs, x='ecart-type', y='notes_moyennes', fill=True, cmap="viridis", cbar=True)

plt.title('Density Heatmap of Mean Rating vs. Standard Deviation for Each authors')
plt.xlabel('Standard Deviation of Ratings')
plt.ylabel('Mean Rating')

plt.grid(True)  # Adding grid for better readability

plt.show()
```

```python
#Tout les auteurs
jeux_uniques = avis["title"].unique()
print(f"nombre de jeux: {len(jeux_uniques)}")
```

```python
# Convertir la liste des auteurs uniques en DataFrame pandas
jeux = pd.DataFrame(jeux_uniques, columns=["title"])

# Ajouter la colonne les colonnes avec des valeurs initiales à zéro
jeux["nb_avis"] = 0
jeux["notes_moyennes"] = np.nan
jeux["note_max"]= np.nan
jeux["note_min"] = np.nan
jeux["ecart-type"] = np.nan
jeux["nb_commentaires"] = 0
# Afficher les premières lignes du DataFrame avec la nouvelle colonne ajoutée
jeux.head(5)
```

```python
jeux = pd.read_csv('../Data/jeux.csv')
jeux.head()
```

```python
# Attempt a direct and efficient plotting method for Mean Rating vs. Standard Deviation

plt.figure(figsize=(10, 6))

# Direct scatter plot using matplotlib for more control and efficiency
plt.scatter(jeux['ecart-type'], jeux['notes_moyennes'], alpha=0.7, color='blue', edgecolor='black')

# Enhancing the plot
plt.title('Mean Rating vs. Standard Deviation for Each Games')
plt.xlabel('Standard Deviation of Ratings')
plt.ylabel('Mean Rating')
plt.grid(True)  # Adding grid for better readability

plt.show()
```

```python
# Définir la taille de la figure
plt.figure(figsize=(10, 6))

# Créer la heatmap de densité
sns.kdeplot(data=jeux,x="ecart-type",y="notes_moyennes",fill=True,thresh=0.01,cmap="viridis",alpha=0.7,cbar=True)

# Ajouter des titres et des labels
plt.title('Density Heatmap of Mean Rating vs. Standard Deviation for Each Games')
plt.xlabel('Standard Deviation of Ratings')
plt.ylabel('Mean Rating')

plt.grid(True)  # Ajouter une grille pour une meilleure lisibilité
plt.show()
```
